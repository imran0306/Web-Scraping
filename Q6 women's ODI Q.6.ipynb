```python
# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
#b) Top 10 women’s ODI Batting players along with the records of their team and rating.
#c) Top 10 women’s ODI all-rounder along with the records of their team and rating.
```

```python
import pandas as pd 
import numpy as np
import requests 
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

# Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
team_string = soup.find_all('span',class_="u-hide-phablet")
team_string
```

```python
team_container = []
for team in team_string:
    team = team.get_text().replace('\n', "")
    team = team.strip(" ")
    team_container.append(team)
team_container
```

```python
#MATCH and  points
match_string = soup.find_all('td',class_="table-body__cell u-center-text")
match_string
```

```python
match_container = []
for match in match_string:
    match = match.get_text().replace('\n', "")
    match = match.strip(" ")
    match_container.append(match)
match_container
```

```python
rate_string = soup.find_all('td',class_="table-body__cell u-text-right rating")
rate_string
```

```python
rate_container = []
for rate in rate_string:
    rate = rate.get_text().replace('\n', "")
    rate = rate.strip(" ")
    rate_container.append(rate)
rate_container
```
